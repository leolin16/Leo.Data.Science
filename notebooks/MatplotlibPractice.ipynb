```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../data/raw/Matplotlib/countries.csv')
data.head()
```

## Line Charts/Time Series

```python
us = data[data.country == 'United States']
plt.plot(us.year, us.gdpPerCapita)
plt.title('GDP Per Capita Growth in the U.S.')
plt.xlabel('year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
cn = data[data.country == 'China']
plt.plot(us.year, us.gdpPerCapita)
plt.plot(cn.year, cn.gdpPerCapita)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita')
plt.show()
```

```python
us.gdpPerCapita.iloc[0] # (integer location)first year of gdppercapita value
```

```python
us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
cn_growth = cn.gdpPerCapita / cn.gdpPerCapita.iloc[0] * 100
plt.title('GDP Per Capita Growth (first year = 100)')
plt.plot(us.year, us_growth)
plt.plot(cn.year, cn_growth)
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('GDP per capita growth')
plt.show()
```

```python
plt.plot(us.year,us.population / 10**6)
plt.plot(cn.year,cn.population / 10**6)
plt.title('Population Growth in the U.S. and China (in millions)')
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show()
```

```python
us_growth = us.population / us.population.iloc[0] * 100
cn_growth = cn.population / cn.population.iloc[0] * 100
plt.plot(us.year, us_growth)
plt.plot(cn.year, cn_growth)
plt.title('Population Growth in the U.S. and China (first year = 100)')
plt.legend(['United States', 'China'])
plt.xlabel('year')
plt.ylabel('population')
plt.show
```

### GDP per capita - life expectancy

```python
years_sorted = sorted(set(data.year))
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.title(given_year)
    plt.xlim(0, 60000)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
    plt.savefig('../reports/figures/' + str(given_year), dpi=200)
    plt.clf()
#     plt.show()
```

```python
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
    plt.title(given_year)
    plt.xlim(2, 5)
    plt.ylim(25, 85)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
    plt.savefig('../reports/figures/log_' + str(given_year), dpi=200)
    plt.clf()
```

### GDP - life expectancy

```python
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(data_year.gdpPerCapita * data_year.population / 10**9, data_year.lifeExpectancy, 5)
    plt.title(given_year)
#     plt.xlim(0, 60000)
#     plt.ylim(25, 85)
    plt.xlabel('GDP in Billions')
    plt.ylabel('Life Expectancy')
    plt.show()
```

```python
for given_year in years_sorted:
    data_year = data[data.year == given_year]
    plt.scatter(np.log10(data_year.gdpPerCapita * data_year.population / 10**9), data_year.lifeExpectancy, 5)
    plt.title(given_year)
#     plt.xlim(0, 60000)
#     plt.ylim(25, 85)
    plt.xlabel('GDP(log10) in Billions')
    plt.ylabel('Life Expectancy')
    plt.show()
```

## Bar Graphs

```python
## good for comparing numeric values
top10 = data_2007.sort_values('population', ascending=False).head(10)
x = range(10)
plt.bar(x, top10.population / 10**6)
plt.xticks(x, top10.country, rotation='vertical')
plt.title('10 most populous countries')
plt.ylabel('Population in Millions')
plt.show()
```

```python
plt.subplot(2,1,1)
x = range(10)
plt.bar(x, top10.population / 10**6)
plt.xticks([], [])
plt.title('10 Most Populous Countries')
plt.legend(['Population in Millions'])
plt.subplot(212)
plt.bar(x, top10.gdpPerCapita * top10.population / 10**9)
plt.xticks(x, top10.country, rotation='vertical')
plt.legend(['GDP in Billions'])
plt.show()
```

```python
# np.arange(10) is similar to range(10), and it allows us to shift
# each value in it by the bar width as you can see below.
x = np.arange(10)

# We need to create subplots in order to overlay two bar plots
# with proper axes on the left hand side and the right hand side.
fig, ax1 = plt.subplots()

width = 0.3 # This is the width of each bar in the bar plot.
plt.xticks(x, top10.country, rotation='vertical')
population = ax1.bar(x, top10.population / 10**6, width)
plt.ylabel('Population')

# ax1.twinx() gives us the same x-axis with the y-axis on the right.
ax2 = ax1.twinx()
gdp = ax2.bar(x + width, top10.gdpPerCapita * top10.population / 10**9,
              width, color='orange')
plt.ylabel('GDP')
plt.legend([population, gdp],
           ['Population in Millions', 'GDP in Billions'])
figure = plt.gcf() # get current figure
plt.show()
```

## Data Aggregation

```python
data_obama = pd.read_csv('../data/raw/matplotlib/obama.csv', parse_dates=['year_month'])
data_obama_mean = data_obama.groupby('year_month').mean()
data_obama_median = data_obama.groupby('year_month').median()
data_obama_25 = data_obama.groupby('year_month').quantile(0.25)
data_obama_75 = data_obama.groupby('year_month').quantile(0.75)
plt.plot(data_obama_mean.index, data_obama_mean.approve_percent, 'red') # index is groupby column since they've become unique
plt.plot(data_obama_median.index, data_obama_median.approve_percent, 'gold')
plt.plot(data_obama_25.index, data_obama_25.approve_percent, 'blue')
plt.plot(data_obama_75.index, data_obama_75.approve_percent, 'green')
plt.legend(['Mean', 'Median', '25th Percentile', '75th Percentile'])
plt.plot(data_obama.year_month, data_obama.approve_percent, 'o', markersize=2, alpha=0.3)
plt.show()
```

### Is Kuwait really the richest country in the world on a per-person basis?

```python
kuwait = data[data.country == 'Kuwait']
plt.plot(kuwait.year, kuwait.gdpPerCapita)
```

```python
plt.subplot(311)
plt.title('GDP Per Capita')
plt.plot(kuwait.year, kuwait.gdpPerCapita)

plt.subplot(312)
plt.title('GDP in Billions')
plt.plot(kuwait.year, kuwait.population*kuwait.gdpPerCapita / 10**9)

plt.subplot(313)
plt.title('Population in Millions')
plt.plot(kuwait.year, kuwait.population / 10**6)

plt.tight_layout()
plt.show()
```

```python
plt.plot(kuwait.year, kuwait.population / kuwait.population.iloc[0]*100)
kuwait_gdp = kuwait.population * kuwait.gdpPerCapita
plt.plot(kuwait.year, kuwait_gdp / kuwait_gdp.iloc[0] * 100)
plt.plot(kuwait.year, kuwait.gdpPerCapita / kuwait.gdpPerCapita.iloc[0] * 100)
plt.title('GDP and Population Growth in Kuwait (first year = 100)')
plt.legend(['Population Growth', 'GDP Growth', 'GDP Per Capita Growth'])
plt.show()
```
