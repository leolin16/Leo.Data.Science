```python
import pandas as pd
import numpy as np
import os
```

## Import Data

```python
raw_data_path = os.path.join(os.path.pardir,'data','raw','titanic')
train_data_path = os.path.join(raw_data_path, 'train.csv')
test_data_path = os.path.join(raw_data_path, 'test.csv')
```

```python
train_df = pd.read_csv(train_data_path, index_col='PassengerId')
test_df = pd.read_csv(test_data_path, index_col='PassengerId')
```

```python
type(train_df)
```

## Basic Structure

```python
train_df.info()
```

```python
test_df.info()
```

```python
## to predict the survival in the test dataset, we first give the default value
test_df['Survived']=-888
```

```python
## concat together
df = pd.concat((train_df, test_df), axis=0, sort=False) # shall pass targets as a tuple, axis=0 means row-wise => stacked over, axis=1 means column-wise => side-by-side
```

```python
## filter rows based on the condition
male_passengers = df.loc[df.Sex == 'male',:]
print('Number of male passengers : {0}'.format(len(male_passengers)))
```

```python
## use & and | to build complex logic
male_passengers_first_class = df.loc[((df.Sex == 'male') & df.Pclass == 1),:]
print('Number of male passengers in first class : {0}'.format(len(male_passengers_first_class)))
```

## Summary Statistics

```python
## description for all numeric columns
df.describe()
```

```python
print('Mean fare : {0}'.format(df.Fare.mean())) # easily affected by extreme values
print('Median fare : {0}'.format(df.Fare.median()))
print('Min fare : {0}'.format(df.Fare.min()))
print('Max fare : {0}'.format(df.Fare.max()))
print('fare range : {0}'.format(df.Fare.max()-df.Fare.min())) # easily affected by extreme values
print('25 percentile of fare : {0}'.format(df.Fare.quantile(.25)))
print('50 percentile of fare : {0}'.format(df.Fare.quantile(.5)))
```
