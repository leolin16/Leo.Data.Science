# Data Preparation

```python
import os
import pandas as pd
raw_path = os.path.join(os.path.pardir,'data','raw')
auto_path = os.path.join(raw_path,'sklearn','imports-85.data')
auto_data = pd.read_csv(auto_path,sep=r'\s*,\s*',engine='python')
auto_data.head()
```

```python
import numpy as np
auto_data = auto_data.replace('?', np.nan)
auto_data.head()
```

```python
auto_data.describe()
```

```python
auto_data.describe(include='all')
```

```python
auto_data['price'].describe() # the type is object
```

```python
auto_data['price'] = pd.to_numeric(auto_data['price'],errors='coerce') # convert column to numeric, ignore errors
auto_data['horsepower'] = pd.to_numeric(auto_data['horsepower'],errors='coerce')
auto_data['bore'] = pd.to_numeric(auto_data['bore'],errors='coerce')
auto_data['stroke'] = pd.to_numeric(auto_data['stroke'],errors='coerce')
auto_data['peak-rpm'] = pd.to_numeric(auto_data['peak-rpm'],errors='coerce')
print(auto_data['price'].describe()) # the type is now float64
print(auto_data['horsepower'].describe())
print(auto_data['bore'].describe())
print(auto_data['stroke'].describe())
print(auto_data['peak-rpm'].describe())
```

```python
auto_data = auto_data.drop('normalized-losses',axis=1)
auto_data.head()
```

```python
cylinders_dict = {'two': 2,
                  'three': 3,
                  'four': 4,
                  'five': 5,
                  'six': 6,
                  'eight': 8,
                  'twelve': 12}
auto_data['num-of-cylinders'].replace(cylinders_dict, inplace=True)
auto_data.head()
```

```python
auto_data = pd.get_dummies(auto_data,columns=['make',
                                              'fuel-type',
                                              'aspiration',
                                              'num-of-doors',
                                              'body-style',
                                              'drive-wheels',
                                              'engine-location',
                                              'engine-type',
                                              'fuel-system']) # convert to one hot representation, since these are discrete categorical values
auto_data.head()
```

```python
auto_data.info()
```

```python
auto_data = auto_data.dropna()
auto_data[auto_data.isnull().any(axis=1)]
```

# Training

```python
from sklearn.model_selection import train_test_split
X = auto_data.drop('price', axis=1)
Y = auto_data['price']
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
linear_model.fit(x_train,y_train)
```

```python
linear_model.score(x_train,y_train)
```

```python
linear_model.coef_
```

```python
predictors = x_train.columns
coef = pd.Series(linear_model.coef_, predictors).sort_values()
coef
```

```python
y_predict = linear_model.predict(x_test)
```

# Lasso & Ridge

```python
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
lasso_model = Lasso(alpha=0.5, normalize=True) # normalize means subtracting the mean and dividing by the L2 norm
ridge_model = Ridge(alpha=0.05, normalize=True)
# alpha = 5 is even better, means more significant variables will drive our prediction
lasso_model.fit(x_train,y_train)
ridge_model.fit(x_train,y_train)
```

```python
print(lasso_model.score(x_train,y_train)) # r square score
print(ridge_model.score(x_train,y_train))
```

```python
predictors = x_train.columns
coef_lasso = pd.Series(lasso_model.coef_, predictors).sort_values()
coef_ridge = pd.Series(ridge_model.coef_, predictors).sort_values()
print(coef_lasso)
print(coef_ridge)
```

```python
y_predict_lasso = lasso_model.predict(x_test)
y_predict_ridge = ridge_model.predict(x_test)
```

```python
%pylab inline
pylab.rcParams['figure.figsize'] = (15,6)
plt.plot(y_predict_lasso, label='Predicted Lasso')
plt.plot(y_predict_ridge, label='Predicted Ridge')
plt.plot(y_test.values, label='Actual')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
r_square_lasso = lasso_model.score(x_test,y_test)
r_square_ridge = ridge_model.score(x_test,y_test)
print(r_square_lasso)
print(r_square_ridge)
```

```python
lasso_model_mse = mean_squared_error(y_predict_lasso,y_test)
ridge_model_mse = mean_squared_error(y_predict_ridge,y_test)
print(math.sqrt(lasso_model_mse))
print(math.sqrt(ridge_model_mse))
```
