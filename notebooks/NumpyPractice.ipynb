```python
# flatten: can flatten arrays into flat sequential
students = np.array([['Alice','Beth','Cathy','Dorothy'],
                     ['65','78','90','81'],
                     ['71','82','79','92']])
for element in students.flatten(order='F'): # order='F' means column major flattening
    print(element)
```

```python
for element in np.nditer(students): #nditer is row major flattening
    print(element)
for element in np.nditer(students, order='F'): #order='F' is column major flattening, others like flags=['external_loop'], op_flags=['readwrite']
    print(element)
    
# flatten returns an array, but nditer just allow iteration
```

```python
## shallow copy => ?? = np.array([]).view(), shallow copy will link the elements within base and derived copy together,but shape is not linked, you can reshape a shallow copy while retaining the identicity of the content
fruits = np.array(["Apple", "Mango", "Grapes", "Watermelon"])
basket_1 = fruits.view()
print(id(fruits))
print(id(basket_1))
print(basket_1 is fruits)
basket_1.base is fruits
```

```python
fruits[0] = "Strawberry"
basket_1
```

```python
## deep copy
basket = fruits.copy()
print(basket is fruits)
print(basket.base is fruits)
```

```python
# broadcasting array - non equal dimension could be broadcasted but should have at least dimension size 1
# default is to match the final(right) dimension first
a = np.random.random((2,3))
print(a)
c = np.random.randn(1,3)
print(c)
a + c
```

```python
## broadcasting Arrays - last(trailing) or one of? dimension should be 1, other parent dimensions should be identical
a=np.array([1,2,3,4,5])
b=np.array([10,10,10,10,10])
print(a*b)
```

```python
def addsubtract(a,b):
    "Return a-b if a > b, otherwise return a + b"
    if a > b:
        return a - b
    else:
        return a + b
```
